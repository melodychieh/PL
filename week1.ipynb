```python
df = pd.read_csv('adult.csv')
```

```python
# 計算薪資的平均值
mean_salary = np.mean(df['Income'])
print("平均薪資:", mean_salary)
```

```python
# 計算薪資的標準差
std_salary = np.std(df['Income'])
print("薪資標準差:", std_salary)
```

```python
# 計算薪資的中位數
median_salary = np.median(df['Income'])
print("薪資中位數:", median_salary)
```
